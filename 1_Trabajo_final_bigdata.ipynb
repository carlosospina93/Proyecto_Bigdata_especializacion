# 1. Caso de Negocio

## Descripción del Problema

Las empresas de telecomunicaciones enfrentan una alta rotación de clientes, fenómeno conocido como **Customer Churn**. La pérdida constante de clientes genera disminución de ingresos, incremento de costos de adquisición de nuevos usuarios y reducción de la rentabilidad del negocio.

Actualmente, la identificación de clientes con riesgo de abandono se realiza de forma reactiva, cuando el cliente ya ha tomado la decisión de cancelar el servicio. Esto limita la capacidad de la organización para implementar estrategias de retención oportunas.

Mediante el análisis de datos masivos y la implementación de modelos predictivos, se busca identificar patrones de comportamiento asociados al abandono de clientes y generar alertas tempranas que permitan diseñar campañas de retención más efectivas.

### Dataset Seleccionado

- IBM Telco Customer Churn Dataset
- Fuente: Kaggle
- Registros: 7.043 clientes
- Variables: Información demográfica, servicios contratados, facturación, satisfacción y abandono.
- Link: https://www.kaggle.com/datasets/yeanzc/telco-customer-churn-ibm-dataset

---

## Objetivo General

Desarrollar una solución Big Data que permita identificar clientes con alta probabilidad de abandono utilizando técnicas de analítica de datos y aprendizaje automático.

---

## Objetivos Específicos

- Centralizar y procesar grandes volúmenes de información de clientes.
- Implementar una arquitectura de datos basada en la metodología Medallion (Bronze, Silver y Gold).

- Analizar las variables que influyen en la cancelación del servicio.

- Generar visualizaciones que apoyen la toma de decisiones estratégicas.

## Impacto Esperado

La solución permitirá:

## Flujo de Información

### Bronze

Almacenamiento de los datos originales provenientes de Kaggle sin modificaciones.

### Silver

Aplicación de reglas de calidad:

- Eliminación de registros duplicados.
- Tratamiento de valores nulos.
- Conversión de tipos de datos.
- Estandarización de formatos.

### Gold

Generación de:

- Indicadores de negocio.
- Métricas de churn.
- Dataset para Machine Learning.
- Dataset para dashboards.

### Machine Learning

Entrenamiento de modelos predictivos para identificar clientes con alta probabilidad de abandono.

### Visualización

Publicación de indicadores estratégicos mediante dashboards interactivos para facilitar la toma de decisiones.
